## Load the MRI images


```python
def load_nii_file(file_path):
    try:
        img = nib.load(file_path)
        return img.get_fdata()
    except Exception as e:
        raise

#dataset path
dataset_path = "BraTS-PEDs2024_Training"

#store data in dict
image_dict_segnii = {}
image_dict_t1cnii = {}
image_dict_t1nnii = {}
image_dict_t2fnii = {}
image_dict_t2wnii = {}
user_count = 0

#iterate through all folders
for folder_name in os.listdir(dataset_path):
    #stop after 25 users, keeping it small for now to test model
    if user_count >= 25: 
        break
    folder_path = os.path.join(dataset_path, folder_name)
    
    if os.path.isdir(folder_path):
        # initialize variables to store file paths
        seg_file = None
        t1c_file = None
        t1n_file = None
        t2f_file = None
        t2w_file = None
        
        # iterate through each file
        for file_name in os.listdir(folder_path):
            if file_name.endswith("seg.nii.gz"):
                seg_file = os.path.join(folder_path, file_name)
            elif file_name.endswith("t1c.nii.gz"):
                t1c_file = os.path.join(folder_path, file_name)
            elif file_name.endswith("t1n.nii.gz"):
                t1n_file = os.path.join(folder_path, file_name)
            elif file_name.endswith("t2f.nii.gz"):
                t2f_file = os.path.join(folder_path, file_name)
            elif file_name.endswith("t2w.nii.gz"):
                t2w_file = os.path.join(folder_path, file_name)
        
        #store data
        if seg_file and t1c_file and t1n_file and t2f_file and t2w_file:
            try:
                print(f"Processing folder: {folder_name}")
                image_dict_segnii[folder_name] = load_nii_file(seg_file)
                image_dict_t1cnii[folder_name] = load_nii_file(t1c_file)
                image_dict_t1nnii[folder_name] = load_nii_file(t1n_file)
                image_dict_t2fnii[folder_name] = load_nii_file(t2f_file)
                image_dict_t2wnii[folder_name] = load_nii_file(t2w_file)
            except Exception as e:
                print(f"Error processing folder {folder_name}: {e}")
        else:
            print(f"Missing files in folder: {folder_name}")
        user_count += 1

print("Data loading complete.")
```

```python
def standardize(data):
    mean = np.mean(data)
    std = np.std(data)
    return (data - mean) / std

# standardize all to ensure features contribute equally
for folder_name in image_dict_t1cnii:
    if folder_name in image_dict_t1cnii:
        image_dict_t1cnii[folder_name] = standardize(image_dict_t1cnii[folder_name])
    if folder_name in image_dict_t1nnii:
        image_dict_t1nnii[folder_name] = standardize(image_dict_t1nnii[folder_name])
    if folder_name in image_dict_t2fnii:
        image_dict_t2fnii[folder_name] = standardize(image_dict_t2fnii[folder_name])
    if folder_name in image_dict_t2wnii:
        image_dict_t2wnii[folder_name] = standardize(image_dict_t2wnii[folder_name])
    
```

```python
multi_channel_data = {}
for folder_name in image_dict_segnii.keys():
    t1c = image_dict_t1cnii[folder_name]
    t1n = image_dict_t1nnii[folder_name]
    t2f = image_dict_t2fnii[folder_name]
    t2w = image_dict_t2wnii[folder_name]
    
    #stack the images
    multi_channel_data[folder_name] = np.stack([t1c, t1n, t2f, t2w], axis=-1)
```
